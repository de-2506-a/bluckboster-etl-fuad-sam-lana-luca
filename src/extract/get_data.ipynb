### Step 2 again: actually use something that works well with pandas to pull data from the db

```python
# create a sql engine
engine = db_engine()

# write a query
query = ''' 
            SELECT 
                film.film_id, 
                film.title, 
                language.name 
            FROM 
                film 
            JOIN 
                language 
            ON 
                film.language_id = language.language_id
            ;         
        '''
# create a pandas dataframe using the engine and query
df = pd.read_sql(query, engine, index_col='film_id')

# print top 25 results
df.head(25)
```
